# 2.数据处理

### 2.1数据格式转换

```python
df.replace('暂无',np.nan,inplace=True)
df['area'] = df['area'].map(lambda x: x.replace('平米','')).astype('float')
df['unit_price'] = df['unit_price'].map(lambda x: x.replace('元/平米','')).astype('float')
def process_year(year):
    if year is not None:
        year = str(year)[:4]
    return year   
df['year'] = df['year'].map(process_year)
# df['year'] = df['year'].map(lambda x: x.replace('n','')).astype('float')
floor = {'低楼层': '低','中楼层': '中','高楼层': '高','低层': '低','中层': '中','高层': '高'}
df['floor'] = df['floor'].map(floor)
def process_area(area):
    if area != '新区':
        area = area.replace('区','').replace('县','')
    return area   
df['district'] = df['district'].map(process_area)
df.replace('nan',np.nan,inplace=True)
df
```

### 2.2重复值处理

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

### 2.3缺失值处理

```python
df.info()
```

```python
df.isnull().sum()[df.isnull().sum()>0]
```

```python
# 户型、朝向、楼层处理，缺失值数量不多，考虑直接删除即可
df.dropna(subset=['layout','facing','floor'], inplace=True)
# 建筑年代、建筑类别、建筑结构、住宅类别、产权性质、装修 这些离散型变量很难处理，得根据实际情况填充，为了得到更加真实的结果暂不处理
# 电梯处理（底层无，高层有，中层随机处理）
df.loc[(df['floor'] == '高') & (df['elevator'].isnull()),'elevator'] = '有 '
df.loc[(df['floor'] == '低') & (df['elevator'].isnull()),'elevator'] = '无 '
df.loc[(df['floor'] == '中') & (df['elevator'].isnull()),'elevator'] = random.choice(['有 ','无 '])

# 分解户型
def apart_room(x):
    room = x.split('室')[0]
    return int(room)
def apart_hall(x):
    hall = x.split('厅')[0].split('室')[1]
    return int(hall)
# def apart_bathroom(x):
#     bathroom = x.split('卫')[0].split('厅')[1]
#     return int(bathroom)
df['room'] = df['layout'].map(apart_room)
df['hall'] = df['layout'].map(apart_hall)
# df['bathroom'] = df['layout'].map(apart_bathroom)
# 删除楼层、户型、单价
df.drop(columns=['layout', 'building', 'res_category'],inplace=True)
# 重新排列字段
columns1 = ['district', 'furnish', 'facing', 'floor', 'room', 'hall', 'elevator', 'bui_category', 'property', 'school', 'year', 'area', 
            'unit_price', 'price']
df = pd.DataFrame(df, columns=columns1)
df = df.dropna().reset_index(drop=True)
df.reset_index(drop=True, inplace=True)

df
```

### 3.6兰州各区域二手房房价平均值柱状图

```python
# 对二手房区域分组对比二手房数量和每平米房价
df_house_count = df.groupby('district')['price'].count().sort_values(ascending=False)
df_price_mean = df.groupby('district')['unit_price'].mean().sort_values(ascending=False)

#设置X轴刻度标签：
def auto_xtricks(rects,xticks):
    x = []
    for rect in rects:
        x.append(rect.get_x() + rect.get_width()/2)
    x = tuple(x)
    plt.xticks(x,xticks)

#设置数据标签：
def auto_tag(rects, data = None, offset = [0,0], size=14):
    for rect in rects:
        try:
            height = rect.get_height()
            plt.text(rect.get_x()+rect.get_width()/2.4, 1.01*height, '%s' % int(height), fontsize=size)
        except AttributeError:
            x = range(len(data))
            y = data.values
            for i in range(len(x)):
                plt.text(x[i]+offset[0],y[i]+0.05+offset[1],y[i], fontsize='14')
                
def auto_tag_float(rects, data = None, offset = [0,0], size=14):
    for rect in rects:
        try:
            height = rect.get_height()
            plt.text(rect.get_x()+rect.get_width()/2.4, 1.01*height, '%s' % round(float(height),1), fontsize=size)
        except AttributeError:
            x = range(len(data))
            y = data.values
            for i in range(len(x)):
                plt.text(x[i]+offset[0],y[i]+0.05+offset[1],y[i], fontsize='14')
```

```python
# 各区域二手房平均价格对比
plt.figure(figsize=(10,5))
plt.rc('font', family='SimHei', size=13)
plt.style.use('ggplot')
# plt.subplot(211)
plt.title(u'各区域二手房平均价格柱状图', fontsize=20)
plt.ylabel(u'二手房平均价格（单位：万/平米）', fontsize=15)
rect2 = plt.bar(np.arange(len(df_price_mean.index)), df_price_mean.values, color='c')
auto_xtricks(rect2, df_price_mean.index)
auto_tag_float(rect2, offset=[-1,0])

plt.show()
```

### 3.7各房价区间的二手房数量统计

```python
# 获取二手房总价的范围
def get_price_range(price, base=50):
    return '{0}-{1}'.format(int(price//base)*base, int(price//base)*base+base)

# 获取二手房面积的范围
def get_area_range(area, base=20):
    return '{0}-{1}'.format(int(area//base)*base, int(area//base)*base+base)
```

```python
# 筛选房屋总价小于50万的二手房信息进行统计
df['groupPrice'] = df['price'].apply(get_price_range)
df['groupPriceSplit'] = df['groupPrice'].str.extract('(\d+)-\d+', expand=False)
df['groupPriceSplit'] = df['groupPriceSplit'].astype('int')
sort_by_price_range = df.loc[df['groupPriceSplit']<210, ['groupPrice','price','groupPriceSplit']]
sort_by_price_range.set_index('groupPrice', inplace=True)
sort_by_price_range.sort_values(by='groupPriceSplit', inplace=True)

# 对房价和房屋面积分组
df_group_price = sort_by_price_range.groupby('groupPrice')['price'].count()

# 房价范围 vs 房屋数量可视化分析
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.title(u'兰州二手房房价/数量统计', fontsize=15)
plt.xlabel(u'二手房房价区间(单位：万)', fontsize=15)
plt.ylabel(u'二手房数量', fontsize=15)
rect_group_price = plt.bar(np.arange(len(df_group_price.index)), df_group_price.values)
auto_xtricks(rect_group_price, df_group_price.index)
auto_tag(rect_group_price, offset=[-1,0])

# 筛选房屋面积小于50的二手房信息进行统计
df['groupArea'] = df['area'].apply(get_area_range)
df['groupAreaSplit'] = df['groupArea'].str.extract('(\d+)-\d+', expand=False)
df['groupAreaSplit'] = df['groupAreaSplit'].astype('int')
sort_by_area_range = df.loc[df['groupAreaSplit']<200, ['groupArea','area','groupAreaSplit']]
sort_by_area_range.set_index('groupArea', inplace=True)
sort_by_area_range.sort_values(by='groupAreaSplit', inplace=True)
# display(sort_by_area_range)

df_group_area = sort_by_area_range.groupby('groupArea')['area'].count()

plt.subplot(122)
plt.title(u'武汉二手房面积/数量统计', fontsize=15)
plt.xlabel(u'二手房房屋面积区间', fontsize=15)
plt.ylabel(u'二手房数量', fontsize=15)
rect_group_area = plt.bar(np.arange(len(df_group_area.index)), df_group_area.values)
auto_xtricks(rect_group_area, df_group_area.index)
auto_tag(rect_group_area, offset=[-1,0])

plt.show()
```

### 网格搜索

```python
# Scikit中使用Grid_Search来获取模型的最佳参数
def grid_search(model, X, y, params):
    grid = GridSearchCV(estimator = model, param_grid = params, scoring = "r2", cv = 5)
    grid.fit(X,y)
    # 指定模型为效果最好参数对应的模型
    best_model = grid.best_estimator_
    # 查看最佳分数和最佳参数
    print("Parameters are {} for the optimal model.".format(grid.best_params_))
    print("Optimal model has best score {:,.10f}.".format(grid.best_score_))
    for i in range(len(grid.cv_results_['mean_test_score'])):
        print(grid.cv_results_['params'][i], "Test score: {:,.10f}".format(grid.cv_results_['mean_test_score'][i]))
#         print(grid.cv_results_['params'][i], "Test score: {:,.10f}".format(grid.score(X, y)))
    return best_model
```
