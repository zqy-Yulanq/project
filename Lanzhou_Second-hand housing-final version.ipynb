# 2. Data processing

### 2.1 Data format conversion

```python
def process_year(year):
    if year is not None:
        year = str(year)[:4]
    return year 
def process_area(area):
    if area != '新区':
        area = area.replace('区','').replace('县','')
    return area
# Replace '暂无' in the data with nan in numpy
df.replace('暂无',np.nan,inplace=True)  
# Delete the units of unit price and area
df['unit_price'] = df['unit_price'].map(lambda x: x.replace('元/平米','')).astype('float')
df['area'] = df['area'].map(lambda x: x.replace('平米','')).astype('float')
# Take the first four digits of the year
df['year'] = df['year'].map(process_year)
# Unify the values of floors as '高', '中', and '低'
floor = {'低楼层': '低','中楼层': '中','高楼层': '高','低层': '低','中层': '中','高层': '高'}
df['floor'] = df['floor'].map(floor)
# Unify districts into '区'
df['district'] = df['district'].map(process_area)
# Replace 'nan' in the data with nan in numpy
df.replace('nan',np.nan,inplace=True)
df
```

### 2.3 Missing value handling

```python
# View information for each feature
df.info()
```

```python
# Check for missing values
df.isnull().sum()[df.isnull().sum()>0]
```

```python
# There are not many missing values in layout, facing, floor, so delete them directly
df.dropna(subset=['layout','facing','floor'], inplace=True)

# furnish, property rights, residential category, building type, building category, year are difficult to deal with and must be filled in 
# according to the actual situation. In order to obtain more real results, they will not be dealt with.

# Elevator('无' on the ground floor, '有' on the upper floor, random processing on the middle floor)
df.loc[(df['floor'] == '高') & (df['elevator'].isnull()),'elevator'] = '有 '
df.loc[(df['floor'] == '低') & (df['elevator'].isnull()),'elevator'] = '无 '
df.loc[(df['floor'] == '中') & (df['elevator'].isnull()),'elevator'] = random.choice(['有 ','无 '])

# Decomposition layout
def apart_room(x):
    room = x.split('室')[0]
    return int(room)
def apart_hall(x):
    hall = x.split('厅')[0].split('室')[1]
    return int(hall)
df['room'] = df['layout'].map(apart_room)
df['hall'] = df['layout'].map(apart_hall)

# Delete layout, building type, residential category
df.drop(columns=['layout', 'building', 'res_category'],inplace=True)
# Rearrange fields
columns1 = ['district', 'furnish', 'facing', 'floor', 'room', 'hall', 'elevator', 'bui_category', 'property', 'school', 'year', 'area', 
            'unit_price', 'price']
df = pd.DataFrame(df, columns=columns1)
# Remove missing values and reset index number
df = df.dropna().reset_index(drop=True)
df
```

### 3.6 Histogram of the average price of second-hand housing in various regions of Lanzhou

```python
# Group the second-hand housing area to compare the quantity and unit price of second-hand housing
df_house_count = df.groupby('district')['price'].count().sort_values(ascending=False)
df_price_mean = df.groupby('district')['unit_price'].mean().sort_values(ascending=False)

# Set the X-axis tick labels：
def auto_xtricks(rects,xticks):
    x = []
    for rect in rects:
        x.append(rect.get_x() + rect.get_width()/2)
    x = tuple(x)
    plt.xticks(x,xticks)

# Set data labels
def auto_tag(rects, data = None, offset = [0,0], size=14):
    for rect in rects:
        try:
            height = rect.get_height()
            plt.text(rect.get_x()+rect.get_width()/2.4, 1.01*height, '%s' % int(height), fontsize=size)
        except AttributeError:
            x = range(len(data))
            y = data.values
            for i in range(len(x)):
                plt.text(x[i]+offset[0],y[i]+0.05+offset[1],y[i], fontsize='14')
                
def auto_tag_float(rects, data = None, offset = [0,0], size=14):
    for rect in rects:
        try:
            height = rect.get_height()
            plt.text(rect.get_x()+rect.get_width()/2.4, 1.01*height, '%s' % round(float(height),1), fontsize=size)
        except AttributeError:
            x = range(len(data))
            y = data.values
            for i in range(len(x)):
                plt.text(x[i]+offset[0],y[i]+0.05+offset[1],y[i], fontsize='14')
```

### Grid Search

```python
# Use Grid_Search to get the best parameters for the model
def grid_search(model, X, y, params):
    grid = GridSearchCV(estimator = model, param_grid = params, scoring = "r2", cv = 5)
    grid.fit(X,y)
    # Specify the model as the model corresponding to the best parameter
    best_model = grid.best_estimator_
    # View the best parameters and best score
    print("Parameters are {} for the optimal model.".format(grid.best_params_))
    print("Optimal model has best score {:,.10f}.".format(grid.best_score_))
    # View the test score of the model with different parameters
    for i in range(len(grid.cv_results_['mean_test_score'])):
        print(grid.cv_results_['params'][i], "Test score: {:,.10f}".format(grid.cv_results_['mean_test_score'][i]))
    return best_model
```
